### 1.Identifying the Best-Selling Product


Exploring the sales data to identify the top-selling products. Understanding which products contribute the most to revenue is crucial for making informed business decisions and optimizing sales strategies. We have grouped the products by the quantity ordered and sorted them in descending order to determine the best-selling products. This information will provide valuable insights into which products are driving the most significant portion of sales, allowing businesses to focus their efforts on what works best.

```python
best_products = Sales.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).reset_index()
best_products.head()
```

This code groups the sales data by the "Product" column, sums the quantities ordered for each product, and then sorts the products in descending order based on the total quantity ordered. 
The result will show the best-selling products at the top of the list.

```python
best_products = Sales.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).reset_index()
top_products = best_products.head(10)  
bars = plt.bar(top_products['Product'], top_products['Quantity Ordered'], color='skyblue', edgecolor='black')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.title('Top 10 Best-Selling Products')
plt.xticks(rotation=90)
for bar in bars:
    bar.set_linewidth(2)
    bar.set_edgecolor('black')
plt.show()
```

### The code is aimed at visualizing the top 10 best-selling products in a dataset.

The **headphones** and other consumer electronics are the most popular products in the *United States*. This suggests that people are increasingly interested in technology and products that can help them stay connected and entertained. The popularity of **AAA batteries** and **charging cables** suggests that people are also looking for products that can help them keep their devices powered up.

# D.Revenue Analysis

```python
Sales['Total Sales'] = Sales['Quantity Ordered'] * Sales['Price Each']
total_sales = Sales['Total Sales'].sum()
print("Total Sales:", total_sales)
```

```python
product_sales = Sales.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
bars = plt.bar(product_sales.index, product_sales.values, color='skyblue', edgecolor='black')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product')
plt.xticks(rotation=90)
for bar in bars:
    bar.set_linewidth(2)
    bar.set_edgecolor('black')
plt.show()
```

The chart shows that folks are happy to spend more cash on gadgets like **iPhones and MacBook Pro laptops**. The popularity of **27-inch 4K gaming monitors** and **34-inch ultrawide monitors** also indicates that people want items that make gaming and entertainment better.
But, not so many folks are buying **LG washing machines and dryers**. This might be because home appliances usually cost more than gadgets, or people may not want to upgrade their home appliances as often as their gadgets.
People are ready to spend on things that keep them connected, entertained, and productive. But they're not as willing to spend on things that aren't necessary for their daily lives.

### 1.Sales Analysis Over Time

```python
sales_by_hour = Sales.groupby('Hour')['Sales'].sum()
fig, axes = plt.subplots(2, 1, figsize=(6, 6))
axes[0].plot(sales_by_hour.index, sales_by_hour.values, color='blue', marker='o', label='Total Sales')
axes[0].set_xlabel('Hour of the Day')
axes[0].set_ylabel('Total Sales')
axes[0].set_title('Total Sales by Hour of the Day')
axes[0].grid(True)
axes[0].legend()
sales_by_month = Sales.groupby('Month')['Sales'].sum()
axes[1].plot(sales_by_month.index, sales_by_month.values, color='green', marker='o', label='Total Sales')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Total Sales')
axes[1].set_title('Total Sales by Month')
axes[1].grid(True)
axes[1].legend()
plt.tight_layout()
plt.show()
```

The chart shows that businesses should market their products and services to consumers all day, not just once a month. They should also think about giving smaller discounts and deals during the day to get people to buy more.
1. Sales are highest around 12pm and 6pm. People like to shop during lunch and dinner breaks.
2. December is the best month for sales. People shop more during the holiday season.
3. Sales are pretty steady all year. People shop throughout the year, not just in specific months.

In summary, the graph shows that businesses should market their stuff all day and all year. They should also think about giving smaller discounts and deals to get people to buy more.

## E.Monthly Sales Trends

```python
sales_by_month = Sales.groupby('Month')['Sales'].sum()
plt.figure(figsize=(10, 6))
plt.plot(sales_by_month.index, sales_by_month.values, marker='o', color='Purple', label='Line Plot')
plt.bar(sales_by_month.index, sales_by_month.values, color='lightcoral', label='Bar Plot')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Sales vs. Month')
plt.grid(True)
plt.legend()
plt.show()
```

The graph suggests that the company is facing some challenges, but it is still possible to maintain its market share and customer base during this period. The company should focus on developing and implementing strategies to address the seasonal decline in sales.
The company is experiencing a seasonal decline in sales. This is likely due to a number of factors, such as the end of the summer vacation season and the start of the school year. Consumers are typically less likely to spend money on discretionary items during this time of year.
The company should focus on maintaining its market share and customer base during this period. This can be done by offering discounts and promotions, and by developing marketing campaigns that target consumers who are still likely to spend money during the fall months.
The company should also analyze its sales data to identify any trends or patterns that may be causing the decline in sales. For example, the company may find that sales are declining in certain product categories or in certain geographic regions. Once the company has identified the root cause of the decline in sales, it can develop strategies to address these issues.

# E.Correlation Heatmap

```python
numerical_columns = ['Quantity Ordered', 'Price Each', 'Month', 'Sales', 'Hour']
correlation_matrix = Sales[numerical_columns].corr()
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="viridis")
plt.title('Correlation Heatmap')
plt.show()
```

The graph shows that there are a number of correlations between different product features. Businesses can use these correlations to improve their pricing and marketing strategies-

1.The **correlation between quantity ordered and price each** is the **strongest correlation**, with a correlation coefficient of **0.9**. This suggests that there is a **strong relationship** between these two features.

2.The **correlation between sales and quantity ordered** is the **second strongest correlation**, with a correlation coefficient of **0.8**. This suggests that there is also a **strong relationship** between these two features.

3.The **correlation between sales and price each** is the **third strongest correlation**, with a correlation coefficient of **0.7**. This suggests that there is a **moderate relationship** between these two features.

4.The **correlation between hour and sales** is the **fourth strongest correlation**, with a correlation coefficient of **0.6**. This suggests that there is a **weak relationship** between these two features.

5.The **correlation between month and sales** is the **weakest correlation**, with a correlation coefficient of **0.5**. This suggests that there is a **very weak relationship** between these two features.
